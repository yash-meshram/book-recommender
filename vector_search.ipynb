```python
# Lets save the tagged description column in text file. Line by line.
# 1 description in 1 Line

books["tagged_description"].to_csv(
    "tagged_description.txt",
    sep = "\n",
    index = False,
    header = False
)
```

```python
# Loading the text
# splitting the text by "\n" seperation

raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()

text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")

documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# creating the vector db

db_books = Chroma.from_documents(
    documents,
    embedding = HuggingFaceEmbeddings()
)
```

```python
query = "A book to teach children about nature."
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.strip('"').split()[0])]
```

```python

def retrieve_sematic_recommendations(
        query: str,
        top_k: int = 5
) -> pd.DataFrame:
    reccomemded_books = db_books.similarity_search(query, k = top_k)

    book_list = []

    for i in range(0, len(reccomemded_books)):
        book_list.append(int(reccomemded_books[i].page_content.strip('"').split()[0]))

    return books[books["isbn13"].isin(book_list)]
```

```python
retrieve_sematic_recommendations("A book to teach children about nature")
```
